```python
# -*- coding: utf-8 -*-
import vk_api
import pandas as pd

def session():
    login, password = '+79506072152', ''
    vk_session = vk_api.VkApi(login, password)

    try:
        vk_session.auth()
    except vk_api.AuthError as error_msg:
        print(error_msg)
        return

    vk = vk_session.get_api()

    """ VkApi.method позволяет выполнять запросы к API. В этом примере
        используется метод wall.get (https://vk.com/dev/wall.get) с параметром
        count = 1, т.е. мы получаем один последний пост со стены текущего
        пользователя.
    """
    return vk

vk = session()
```

```python
import numpy as np
post_id_list=set([])
urls=df1['Идентификатор основной статьи'].values
for url in urls:
    if type(url)==str:
        pos=url.find('wall-25106701_')+14
        post_id_list.add(int(url[pos:pos+6]))
        
post_id_list
```

```python
def request(owner_id,post_id):
    response = vk.wall.getComments(owner_id=-owner_id, post_id=post_id, count=100, sort='asc')  # , offset=0
    if response['items']:
        com=pd.DataFrame(response['items'])
        com['Intent']=['']*com.shape[0]
        com['Content']=['']*com.shape[0]
        com['Intervention']=['']*com.shape[0]
#         com['text']=com['text'].apply(lambda a: a.encode('ascii', 'ignore'))
#         print(com['text'])
#         com.to_csv('./interv/%s_%s.csv'%(owner_id, post_id), encoding='ascii')#, encoding='ascii')
        writer = pd.ExcelWriter('./interv/%s_%s.xlsx'%(owner_id, post_id))
        com.to_excel(writer,'Sheet1', encoding='ascii')
```

```python
for post_id in post_id_list:
    try:
        request(25106701,post_id)
    except Exception as e:
        print(post_id, e)
```
